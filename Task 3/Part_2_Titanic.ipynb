# Titanic Dataset Analysis

```python
import pandas as pd

# Load the dataset
titanic = pd.read_csv('Titanic-Dataset.csv')
```

## Step 1: Exploration 

```python
# 1.1 First 5 rows
print("\nFirst 5 rows:")
titanic.head()
```

```python
# 1.2 Dataset information
print("\nDataset Information:")
titanic.info()
```

```python
# 1.3 Statistical summary
print("\nSummary Statistics:")
titanic.describe()
```

## Step 2: Data Cleaning 

### 1. Check missing values before cleaning

```python
print("Missing values before cleaning:")
titanic.isnull().sum()
```

```python
# 2.1 Fill Age with median (middle value - avoids outlier effect)
age_median = titanic['Age'].median()
titanic['Age'] = titanic['Age'].fillna(age_median)
print(f"\nAge filled with median: {age_median}")
```

```python
# 2.2 Fill Embarked with mode (most common value)
embarked_mode = titanic['Embarked'].mode()[0]  # [0] gets the first mode
titanic['Embarked'] = titanic['Embarked'].fillna(embarked_mode)
print(f"Embarked filled with mode: {embarked_mode}")
```

```python
# 2.3 Drop Cabin column (too many missing values)
titanic = titanic.drop('Cabin', axis=1)
```

```python
# 2.4 Remove duplicates if any
duplicates_before = titanic.duplicated().sum()
titanic = titanic.drop_duplicates()
print(f"Duplicates removed: {duplicates_before}")
```

```python
# Check missing values after cleaning
titanic.isnull().sum()
```

### Step 3: Data Analysis Using Groupby

```python
# 3.1 Survival rate by gender
print("\nSurvival Rate by Gender:")
survival_gender = titanic.groupby('Sex')['Survived'].mean() * 100
survival_gender
```

```python
# 3.2 Survival rate by class
print("\nSurvival Rate by Passenger Class:")
survival_class = titanic.groupby('Pclass')['Survived'].mean() * 100
survival_class.round(2)
```

```python
# 3.3 Average age per class
print("\nAverage Age by Passenger Class:")
avg_age_class = titanic.groupby('Pclass')['Age'].mean()
avg_age_class.round(1)
```

```python
# 3.4 Survival rate by age group

# Create Age Group and find survival rate
titanic['AgeGroup'] = pd.cut(titanic['Age'], 
                             bins=[0, 12, 18, 35, 60, 100], 
                             labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])

print("\nSurvival Rate by Age Group:")

survival_age_group = titanic.groupby('AgeGroup', observed=False)['Survived'].mean() * 100
survival_age_group.round(2)
```

### Step 4: Filtering

```python
# 4.1 Female passengers who survived
female_survived = titanic[(titanic['Sex'] == 'female') & (titanic['Survived'] == 1)]
print(f"\nNumber of female passengers who survived: {len(female_survived)}")
female_survived[['Name', 'Age', 'Pclass']].head(10)
```

```python
# 4.2 Children who survived (age < 13)
children_survived = titanic[(titanic['Age'] < 13) & (titanic['Survived'] == 1)]
print(f"Number of children who survived: {len(children_survived)}")
children_survived[['Name', 'Age', 'Sex', 'Pclass']].head(10)
```

```python
# 4.3 Passengers in 1st class with high survival probability

first_class_survived = titanic[(titanic['Pclass'] == 1) & (titanic['Survived'] == 1)]
print(f"\nNumber of 1st class passengers who survived: {len(first_class_survived)}")
print("First 5 examples:")
first_class_survived[['Name', 'Age', 'Sex']].head()
```
